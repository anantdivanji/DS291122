## sklearn preprocessing - One Hot Encoder Function

```python
data = sns.load_dataset('tips')
data.head()
```

```python
## Step1 : To import the library

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe
```

```python
## Step 2 : Apply ohe on text columns of the dataset and store the output in data_ohe

data_ohe = ohe.fit_transform(data[['sex', 'smoker', 'day', 'time']])
data_ohe
```

```python
## Step 3: Convert the output of step to array using .toarray()

data_ohe.toarray()
```

```python
## Step 4 : To get more meaningful output, convert the array from step 3 to dataframe

dataframe_ohe = pd.DataFrame(data_ohe.toarray(), columns = ohe.get_feature_names_out())
dataframe_ohe
```

```python
## Step 5 : Join original data and dataframe_ohe obtained from step 4

data = pd.concat([data, dataframe_ohe], axis = 1)
data
```

```python
## Step 6 : Delete the text from after step5 

data = data.drop(['sex', 'smoker', 'day', 'time'], axis = 1)
data
```

```python
X = data.drop('tip', axis = 1)
y = data['tip']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(X, y)
```
