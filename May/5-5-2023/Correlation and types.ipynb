## 3 Methods to check relation between dependent and independent variable

1. Scatterplot method
2. Regplot method
3. Correlation method

## Loading the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Create a dataset

```python
d1 = pd.DataFrame({
                    'yrs_exp' : [0,2, 4, 5, 7, 8, 10, 12, 15],
                    'salary' : [40000, 49000, 58000, 62000, 79000, 82000, 125000, 160000, 200000]
})
d1
```

## Relation between salary(dep variable) and yrs_exp(indep variable)

```python
sns.scatterplot(x = d1['yrs_exp'], y = d1['salary'])
```

```python
sns.regplot(x = d1['yrs_exp'], y = d1['salary'])
```

## 3. Correlation method

```python
sns.heatmap(d1.corr(), annot = True)
```

## Multiple forms of Relations

```python
d2 = pd.DataFrame({
                    'Sales' : [150, 200, 325, 450, 575, 725],
                    'Profit' : [950, 825, 750, 600, 525, 450]
})
d2
```

```python
sns.regplot(x = d2['Sales'], y = d2['Profit'])
```

```python
sns.heatmap(d2.corr(), annot = True)
```

## Note: 

1. The range of correlation is between -1 to 1.
2. If the correlation is >=  0.5 and near to 1, then we say that the variables are highly positively correlated
3. If the correlation is <= -0.5 and near to -1, then we say that the varaibles are highly negatively correlated
4. IF the correlation is < 0.5 and near to 0, then we say that the varaibles are not correlated

## When to apply linear regression algorithm

1. There has to be a relation between dep and indep variable. The relation can be either positive or negative
2. The relation must be linear in nature
