## Load the standard standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the data

```python
data = pd.read_csv('Movie_regression.csv')
data.head()
```

```python
data.columns
```

```python
data.shape
```

## Missing Value Analysis

```python
data.isnull().sum()
```

```python
data['Time_taken'] = data['Time_taken'].fillna(150)
data.isnull().sum()
```

## Encoding

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe
```

```python
data_ohe = ohe.fit_transform(data[['3D_available', 'Genre']])
data_ohe
```

```python
data_ohe.toarray()
```

```python
data_ohe = pd.DataFrame(data_ohe.toarray(), columns = ohe.get_feature_names_out())
data_ohe
```

```python
data = pd.concat([data, data_ohe], axis = 1)
data.head()
```

```python
data = data.drop(['3D_available', 'Genre'], axis = 1)
data.head()
```

## Seperate X and y

```python
X = data.drop('Collection', axis = 1)
y = data['Collection']
```

## Split the data into train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

## Fit Decision Tree Regressor on X_train and y_train data

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()    ## by default gini index method is used
dtr
```

```python
dtr.fit(X_train, y_train)
```

## Perform Predictions on X_test

```python
y_pred = dtr.predict(X_test)
y_pred
```

```python
X_test
```

## Accuracy Check

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

## Apply Decision Tree Entropy method on the data

```python
from sklearn.tree import DecisionTreeRegressor
dtre = DecisionTreeRegressor(criterion = 'entropy')    
dtre
```

```python
dtre.fit(X_train, y_train)
```

## Observation:

1. Decision Tree with entropy cannot be apply on numerical dependent variable
2. On numerical dependent variable, we can only apply Decision Tree with gini index
