## Load the standard libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Create a data

```python
data = pd.DataFrame({
                    'No of hours studied' : [2,4,6,8,10, 2.5, 3.5, 4.5, 5.5, 6.5],
                    'result' : [0, 1, 0, 1, 1, 0, 0, 1, 1, 1]
})
data
```

## Observations:

1. result column depends on the number of hours studied by the candidate
2. result variable is the target variable
3. No. of hours studied is a independent variable
4. result is a categorical varaible.

```python
sns.scatterplot(x = data['No of hours studied'], y = data['result'])
```

```python
sns.regplot(x = data['No of hours studied'], y = data['result'])
```
