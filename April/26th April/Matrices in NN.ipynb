```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Load the sun flower image

```python
from PIL import Image, ImageOps

img = np.array(Image.open('sun.jpg'))

plt.figure(figsize = (8,8))
plt.imshow(img)
```

## Rotating the img

```python
rot_img = np.rot90(img)
plt.imshow(rot_img)
```

## Negative of the image

```python
img_grey = img.sum(2) / (255 * 3)  ## summing over the 2nd axis (channel axis) to get the grey image
plt.imshow(img_grey)
```

## Greyed image and negative img

```python
img_grey = 255 * 3 - img_grey   ### 255 * 3 because we added along the channel axis in the previous code

fig = plt.figure(figsize = (10, 10))
fig.add_subplot(1,2,1)
plt.imshow(img_grey)
plt.title('Negative of the gray image')

img = 255 - img
fig.add_subplot(1,2,2)
plt.imshow(img)
plt.title('Negative of the orignal image')
```

## Visualizing the Red, Green and Blue Channels of the image

```python
img = np.array(Image.open('sun.jpg'))
img.shape
```

```python
plt.imshow(img)
```

## Note: Red = 0, Green = 1, Blue =2

```python
img_R, img_G, img_B = img.copy(), img.copy(), img.copy()

img_R[:, :, (1,2)] = 0
img_G[:, :, (0,2)] = 0
img_B[:, :, (0,1)] = 0

img_rgb = np.concatenate((img_R, img_G, img_B), axis =1)

plt.figure(figsize = (15, 15))
plt.imshow(img_rgb)
```
